```python
from sentence_transformers import SentenceTransformer
from neo4j import GraphDatabase
from IPython.display import display
from IPython.display import Markdown
```

```python
df = pd.read_csv('/content/drive/MyDrive/translate_receitas_dataset_atualizado2.csv')
df
```

```python
from google.colab import userdata
uri_neo4j = userdata.get('Banco_neo4j')
key_neo4j = userdata.get('neo4j_key')
```

```python
uri = uri_neo4j
user = "neo4j"
password = key_neo4j

try:
    driver = GraphDatabase.driver(uri, auth=(user, password))
    driver.verify_connectivity()
    print("Connected successfully!")
except Exception as e:
    print(f"Connection error: {e}")
```

```python
def compile_text(x):

    text = f"""{x['Recipe Name']},
               {x['Ingredients']},
               {x['Preparation Method']}
            Portion = {x["Portion"]},
            Calories = {x["Calories"]}
            """

    return text
```

```python
sentences = df.apply(lambda x: compile_text(x), axis=1).tolist()
sentences
```

```python
modelB = SentenceTransformer("all-MiniLM-L6-v2")
output = modelB.encode(sentences=sentences,
         show_progress_bar=True,
         normalize_embeddings=True)

embeddings = pd.DataFrame(output)
embeddings
```

```python
df["Embedding"] = embeddings.values.tolist()
df
```

```python
df.info()
```

```python
def create_nodes_and_relationships(df):
    with driver.session() as session:
        session.run("MATCH (n) DETACH DELETE n")

        for _, row in df.iterrows():
            session.run(
                """
                MERGE (r:Recipe {
                    name: $name,
                    preparation_time: $preparation_time,
                    portion: $portion,
                    yield: $yield_,
                    ingredients: $ingredients,
                    preparation_method: $preparation_method,
                    calories: $calories,
                    proteins: $proteins,
                    carbohydrates: $carbohydrates,
                    fats: $fats,
                    fibers: $fibers,
                    dietary_restrictions: $dietary_restrictions
                })
                MERGE (em:Embedding {embedding: $embedding})
                MERGE (em)-[:VETOR_DENSO_DE]->(r)

                """,
                name = row["Recipe Name"],
                preparation_time = row["Preparation Time"],
                portion = row["Portion"],
                yield_ = row["Yield"],
                ingredients = row["Ingredients"],
                preparation_method = row["Preparation Method"],
                calories = int(row["Calories"]),
                proteins = int(row["Proteins"]),
                carbohydrates = int(row["Carbohydrates"]),
                fats = int(row["Fats"]),
                fibers = int(row["Fibers"]),
                dietary_restrictions = row["Dietary Restrictions"],
                embedding = row["Embedding"]
            )
        print('Nodos e relacionamentos criados com sucesso!')
```

```python
df.fillna('')
```

```python
def clean_nan_values(df):
    return df.fillna("")
df = clean_nan_values(df)
```

```python
create_nodes_and_relationships(df)
```

```python
def get_embedding(query):
    return modelB.encode(query).tolist()
```

```python
def search_similar_products(query):
    query_embedding = get_embedding(query)
    with driver.session() as session:
        result = session.run(
            """
            WITH $query_embedding AS queryVec
            MATCH (em:Embedding)-[:VETOR_DENSO_DE]->(r:Recipe)
            WITH r, vector.similarity.cosine(queryVec, em.embedding) AS similarity
            WHERE similarity >= 0.4
            RETURN r.name AS Recipe, similarity
            ORDER BY similarity DESC
            LIMIT 5
            """,
            query_embedding=query_embedding
        )
        return result.data()
```

```python
sentence = "I have eggplant, zucchini, and tomato. What can I make with these ingredients?"
similar_products = search_similar_products(sentence)
similar_products
```

```python
sentence = "I am on a diet and would like a recipe with less than 400 calories"
similar_products = search_similar_products(sentence)
similar_products
```
